```python
import numpy as np
import cv2
from PIL import Image
import os
import pickle
```

```python
def showImg(img):
    # Show the Image
    cv2.imshow('image',img)
    cv2.waitKey(0)
```

```python
def append_img(list, img):
    new_img = img.reshape((-1,1))
    list.append(new_img)
    return list

# load the images
direcotry = 'train'
new_directory = 'data_subset'
label_file = '../ResNet_Model/labels.txt'
f = open(label_file, 'w')
image_size = 128
class_count_list = list()

for category in os.listdir(direcotry):
    class_list = list()
    path = os.path.join(direcotry, category)
    length  = len(os.listdir(path))
    print(length)
    for j in os.listdir(path):
        img = cv2.imread(path+'/'+ j, 0)
        img = cv2.resize(img,(image_size,image_size),interpolation=cv2.INTER_AREA)  # INTER_LINEAR / INTER_CUBIC / INTER_AREA
        img = img.reshape((-1,1))
        class_list.append(img)
        
        if (length < 100):
            # flip 
            horizontal_img = cv2.flip(img, 0)
            class_list = append_img(class_list, horizontal_img)
            vertical_img = cv2.flip(img, 1)
            class_list = append_img(class_list, vertical_img)
            both_img = cv2.flip(img, -1)
            class_list = append_img(class_list, both_img)
            # rotation 
            (h, w) = img.shape[:2]
            center = (w / 2, h / 2)
            M = cv2.getRotationMatrix2D(center, 90, 1.0)
            rotated90 = cv2.warpAffine(img, M, (h, w))
            class_list = append_img(class_list, rotated90)
            M = cv2.getRotationMatrix2D(center, 180, 1.0)
            rotated180 = cv2.warpAffine(img, M, (h, w))
            class_list = append_img(class_list, rotated180)
            M = cv2.getRotationMatrix2D(center, 270, 1.0)
            rotated270 = cv2.warpAffine(img, M, (h, w))
            class_list = append_img(class_list, rotated270)
            # Gaussian filter
            blur5 = cv2.GaussianBlur(img,(5,5),0)
            class_list = append_img(class_list, blur5)
            blur10 = cv2.GaussianBlur(img,(7,7),0)
            class_list = append_img(class_list, blur10)
            blur15 = cv2.GaussianBlur(img,(9,9),0)
            class_list = append_img(class_list, blur15)
            blur20 = cv2.GaussianBlur(img,(13,13),0)
            class_list = append_img(class_list, blur20)
            
        
    print(len(class_list))
    class_count_list.append(len(class_list))
    class_list = np.array(class_list)
    class_list = class_list.reshape((-1,image_size*image_size))
    filename = category + '.npy'
    np.save(new_directory + '/' + filename, class_list)
    f.write(str(len(class_list)) + '\n')
    f.write(category + '\n')
    print(category)
f.close()
```
