```python
import numpy as np
import cv2
from PIL import Image
import os
```

```python
def showImg(img):
    # Show the Image
    cv2.imshow('image',img)
    cv2.waitKey(0)
```

```python
# load the images
img_size = 128
direcotry = './dataset/train'
new_directory = './dataset/data_subset'
label_file = './dataset/labels.txt'
f = open(label_file, 'a+')

for category in os.listdir(direcotry):
	class_list = list()
	path = os.path.join(direcotry, category)
	count = 0
	for j in os.listdir(path):
		count += 1 
		img = cv2.imread(path+'/'+ j, 0)
		img = cv2.resize(img,(img_size, img_size),interpolation=cv2.INTER_AREA)  # INTER_LINEAR / INTER_CUBIC / INTER_AREA 
		img = img.reshape((-1,1))
		class_list.append(img)
	class_list = np.array(class_list)
	class_list = class_list.reshape((-1,img_size * img_size))
	filename = category + '.npy'
	np.save(new_directory + '/' + filename, class_list)
	f.write(str(count) + '\n')
	f.write(category + '\n')
	print(category)
f.close()
```
